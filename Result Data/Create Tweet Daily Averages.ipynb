```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
tweet_gen = pd.read_csv('senti_general_covid_tweet_data.csv',parse_dates=['Datetime'],index_col=['Datetime'])
tweet_ny_gen = pd.read_csv('senti_nyc_popular_tweet_data.csv',parse_dates=['Datetime'],index_col=['Datetime'])
tweet_ny_covid = pd.read_csv('senti_nyc_covid_tweet_data.csv',parse_dates=['Datetime'],index_col=['Datetime'])

tweet_gen = pd.read_csv('senti_general_covid_tweet_data.csv')
tweet_ny_gen = pd.read_csv('senti_nyc_popular_tweet_data.csv')
tweet_ny_covid = pd.read_csv('senti_nyc_covid_tweet_data.csv')

tweet_gen = tweet_gen.drop(columns=['Unnamed: 0','Unnamed: 0.1']);
tweet_ny_gen = tweet_ny_gen.drop(columns=['Unnamed: 0','Unnamed: 0.1']);
tweet_ny_covid = tweet_ny_covid.drop(columns=['Unnamed: 0','Unnamed: 0.1']);
```

```python
tweet_ny_gen.head()
```

```python
#first lets figure out if the spell correction actually makes a difference for the sentiment

tweet_gen.sort_index(inplace=True)
tweet_ny_gen.sort_index(inplace=True)
tweet_ny_covid.sort_index(inplace=True)

mTT = pd.merge(left=tweet_ny_covid,right=tweet_ny_gen,left_index=True,right_index=True)
nEqual = 0

mTT['equal'] = np.where(mTT['Text_x']==mTT['Text_y'],1,0)



print('result similarity %')
print(mTT['equal'].sum()/len(tweet_gen))

# looks like it didn't make much of a difference at all
```

```python
mTT.head()
```

```python
# changing the pos/neg scores to -1 or 1

def normalizeSentiment(score):
    if (score == "[4]"):
        return 1
    else: 
        return -1
    
tweet_gen['senti'] = tweet_gen['senti'].apply(lambda x : normalizeSentiment(x))
tweet_ny_gen['senti'] = tweet_ny_gen['senti'].apply(lambda x : normalizeSentiment(x))
tweet_ny_covid['senti'] = tweet_ny_covid['senti'].apply(lambda x : normalizeSentiment(x))
```

```python
tweet_gen.head()
```

```python

# let's also make a "weighted" column that multiplies by the number of retweets
tweet_gen['weightedSenti'] = tweet_gen['senti'].multiply(tweet_gen['retweets'])
tweet_ny_gen['weightedSenti'] = tweet_ny_gen['senti'].multiply(tweet_ny_gen['retweets'])
tweet_ny_covid['weightedSenti'] = tweet_ny_covid['senti'].multiply(tweet_ny_covid['retweets'])
```

```python
# and a rolling sum column, which we will also weight
tweet_gen['rolling_sum'] = tweet_gen['senti'].rolling(3).sum()
tweet_gen['rolling_sum_weighted'] = tweet_gen['weightedSenti'].rolling(3).sum()

tweet_ny_gen['rolling_sum'] = tweet_ny_gen['senti'].rolling(3).sum()
tweet_ny_gen['rolling_sum_weighted'] = tweet_ny_gen['weightedSenti'].rolling(3).sum()

tweet_ny_covid['rolling_sum'] = tweet_ny_covid['senti'].rolling(3).sum()
tweet_ny_covid['rolling_sum_weighted'] = tweet_ny_covid['weightedSenti'].rolling(3).sum()
```

```python
tweet_gen.head()
```

```python
# should now try to get daily averages 

def dailyAverage(df):
    df = df.drop(columns=['Text','cleanText','retweets','spellCorrected','sentiSC','equal'])
    #print(df)
    #df['Datetime'] = df['Datetime'].dt.date
    #print(type(df['Datetime'][1]))
    df = df.resample('D').mean()
    return df
    
tDaily_gen = dailyAverage(tweet_gen)
tDaily_ny_gen = dailyAverage(tweet_ny_gen)
tDaily_ny_covid = dailyAverage(tweet_ny_covid)

tDaily_gen.head()
tDaily_ny_gen.head()
tDaily_ny_covid.head()
```

```python
tDaily_gen.to_csv('tDaily_gen.csv')
tDaily_ny_gen.to_csv('tDaily_ny_gen.csv')
tDaily_ny_covid.to_csv('tDaily_ny_covid.csv')
```
