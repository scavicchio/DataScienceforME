### Get json file using NYT API request

#### NYT api only allows to request monthly article; therefore each request generate 10 articles for each month. eg: 20200121-20200221; In this notebook, time will be set as three month begining at 20200121 and ending at 20200421. Each request only generates 10 articles which is not diverse enough. To make articles more diverse, I request articles from different page. (similar to clicking load more)

```python
# This cell gets articles for the three month on page 1
r1 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200121&end_date=20200221&page=1&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article1 = r1.json()

r2 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200221&end_date=20200321&page=1&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article2 = r2.json()

r3 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200321&end_date=20200421&page=1&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article3 = r3.json()
```

```python
# This cell gets articles for the three month on page 2 (click load more by 2 times)
r11 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200121&end_date=20200221&page=2&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article11 = r11.json()

r22 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200221&end_date=20200321&page=2&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article22 = r22.json()

r33 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200321&end_date=20200421&page=2&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article33 = r33.json()
```

```python
# This cell gets articles for the three month on page 3 (click load more by 3 times)
r111 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200121&end_date=20200221&page=5&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article111 = r111.json()

r222 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200221&end_date=20200321&page=5&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article222 = r222.json()

r333 = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=coronavirus&begin_date=20200321&end_date=20200421&page=5&sort=relevance&api-key=2WCFSb6GFArnrb10RQGSyBF3D5Wc1SbE")
article333 = r333.json()
```

```python
def parse_articles(articles):
    '''
    This function takes in a response to the NYT api and parses
    the articles into a list of dictionaries
  
    '''
    news = []
    for i in articles['response']['docs']:
        dic = {}
        dic['id'] = i['_id']
        if i['abstract'] is not None:
            dic['abstract'] = i['abstract']
        dic['headline'] = i['headline']['main']
        
        dic['lead_paragraph'] = i['lead_paragraph']
        dic['date'] = i['pub_date'][0:10] # cutting time of day.
        dic['section'] = i['section_name']
        dic['source'] = i['source']
        
        dic['url'] = i['web_url']
        dic['word_count'] = i['word_count']
        
        news.append(dic)
    return(news) 
```

```python
article3 = parse_articles(article3)
article2 = parse_articles(article2)
article1 = parse_articles(article1)
article33 = parse_articles(article33)
article22 = parse_articles(article22)
article11 = parse_articles(article11)
article333 = parse_articles(article333)
article222 = parse_articles(article222)
article111 = parse_articles(article111)
```

```python
# put all dictionaries together in one
article = article1 + article2 + article3 + article33 + article22 + article11 +article111 +article222 +article333
```

```python
article_df = pd.DataFrame(article)
article_df['date'] = pd.to_datetime(article_df['date'])
```

```python
article_df = article_df.sort_values(by="date")
```
