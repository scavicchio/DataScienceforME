```python
# import the 3 tweets 
import pandas as pd
df_gen = pd.read_csv("../Raw Tweets/general_covid_tweet_data.csv")
df_nyc_covid = pd.read_csv("../Raw Tweets/nyc_covid_tweet_data.csv")
df_nyc_gen = pd.read_csv("../Raw Tweets/nyc_popular_tweet_data.csv")

# making sure the tweets are text
df_gen['Text'] = df_gen['Text'].astype(str)
df_nyc_covid['Text'] = df_gen['Text'].astype(str)
df_nyc_gen['Text'] = df_gen['Text'].astype(str)

# cleaning columns I don't think we will need
df_gen = df_gen.drop(columns=['Unnamed: 0','user']);
df_nyc_covid = df_nyc_covid.drop(columns=['Unnamed: 0','user']);
df_nyc_gen = df_nyc_gen.drop(columns=['Unnamed: 0','user'])

# adding a new column to save the processed text
df_gen['cleanText'] = df_gen['Text'].astype(str)
df_nyc_covid['cleanText'] = df_gen['Text'].astype(str)
df_nyc_gen['cleanText'] = df_gen['Text'].astype(str)


df_gen.head()
```

```python
# now I want to do some text cleaning in the actual data
# these are some helper functions that I will use later
import string
def removePunctuation(text):
    # need to do this as would rather have two words than a single
    # word that is supposed to be hyphenated
    text = text.replace('-',' ')
    return "".join([x for x in text if x not in string.punctuation])

import re
def removeMentionsAndURLs(text):
    mentions = r'@[A-Za-z0-9_]+'
    url1 = r'https?://[^ ]+'
    url2 = r'www.[^ ]+' 
    text = re.sub(mentions,'',text)
    text = re.sub(url1,'',text)
    text = re.sub(url2,'',text)
    return str(text)

def makeLowercase(text):
    return text.lower()
    
import nltk
from nltk.corpus import stopwords
def removeStopwords(text): 
    return  ' '.join([x for x in text.split() if x not in stopwords.words('english')])

from textblob import TextBlob
def correctSpelling(text): 
    return TextBlob(text).correct() # about 70% accurate

# this doesn't really work yet
from pycontractions import Contractions
cont = Contractions('resources/GoogleNews-vectors-negative300.bin')
def expandContractions(text): 
    return list(cont.expand_texts(text))
```

```python
# apply the cleaning functions in a speific order
# this takes a long fucking time
def cleanTweetText(text):
    text = removeMentionsAndURLs(text)
    text = correctSpelling(text) #works
    text = makeLowercase(text) #works
    text = removePunctuation(text) #works
    text = removeStopwords(text) #works
    #text = expandContractions(text) # doesnt work yet - womp womp
    return text

print("first dataframe")
df_gen['cleanText'] = df_gen['cleanText'].apply(lambda x : cleanTweetText(x))
print("second dataframe")
df_nyc_covid['cleanText'] = df_nyc_covid['cleanText'].apply(lambda x : cleanTweetText(x))
print("third dataframe")
df_nyc_gen['cleanText'] = df_nyc_gen['cleanText'].apply(lambda x : cleanTweetText(x))
```

```python
df_gen.to_csv('cleaned_general_covid_tweet_data'.csv)
df_nyc_covid.to_csv('cleaned_nyc_covid_tweet_data'.csv)
df_nyc_gen.to_csv('cleaned_nyc_popular_tweet_data'.csv)
```
